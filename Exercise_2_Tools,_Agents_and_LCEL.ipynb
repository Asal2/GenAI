```python
from langchain.prompts import PromptTemplate
from langchain.output_parsers import PydanticOutputParser
from langchain.chains import LLMChain
```

```python
from typing import Optional
from typing import List
from pydantic import BaseModel, Field, field_validator
from datetime import datetime

class CompanyInfo(BaseModel):
    company_name: str = Field(description="The name of the company")
    founding_date: str = Field(description="The founding date in YYYY-MM-DD format, or YYYY if only year available")
    founders: List[str] = Field(description="List of company founders")

    @field_validator("founding_date", mode="before")
    def normalize_date(cls, date_str):
        """
        Normalize date string:
          - YYYY → YYYY-01-01
          - YYYY-MM → YYYY-MM-01
          - YYYY-MM-DD → unchanged
        """
        if not date_str:
            return date_str

        try:
            if len(date_str) == 10:  # YYYY-MM-DD
                datetime.strptime(date_str, "%Y-%m-%d")
                return date_str
            elif len(date_str) == 7:  # YYYY-MM
                datetime.strptime(date_str, "%Y-%m")
                return f"{date_str}-01"
            elif len(date_str) == 4:  # YYYY
                datetime.strptime(date_str, "%Y")
                return f"{date_str}-01-01"
            else:
                return date_str
        except:
            return date_str

class CompanyList(BaseModel):
    companies: List[CompanyInfo] = Field(description="A list of company information objects")
```

```python
from langchain.prompts import PromptTemplate
from langchain.output_parsers import PydanticOutputParser
import json

# Create a parser instance for the list of companies
parser = PydanticOutputParser(pydantic_object=CompanyList)



# Create a prompt template
prompt = PromptTemplate(
    template="""
Extract ALL companies from the text below.

Return valid JSON in this format:
{format_instructions}

Text:
{paragraph}
""",
    input_variables=["paragraph"],
    partial_variables={"format_instructions": parser.get_format_instructions()}
)
```

```python
llm = get_llm("gpt-4.1")  # Use gpt-4.1 instead of nano for better accuracy

# Chain: prompt -> llm -> parser
chain = prompt | llm | parser


results = []
for para in paragraphs:
    if not para.strip():  # skip empty chunks
        continue
    try:
        parsed_output = chain.invoke({"paragraph": para})
        results.append(parsed_output)
    except Exception as e:
        print("Parsing failed:", e)

# 6. Print JSON
import json
for i, r in enumerate(results):
    print(f"\n--- Extraction from Paragraph {i+1} ---")
    if r.companies:
        # Convert the entire CompanyList object to JSON
        # json_output = json.dumps(r.dict(), indent=2)
        # print(json_output)

        dict_output = r.model_dump()
        print("these are dict output")
        print(dict_output)

    else:
        print("  No company information found in this paragraph.")
```
